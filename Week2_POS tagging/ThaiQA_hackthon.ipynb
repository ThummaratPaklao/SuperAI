```python
import pandas as pd
```

```python
wiki_data=pd.read_csv("/content/drive/MyDrive/NLP/hackathon/ThaiQACorpus-DevelopmentDataset.csv")
```

```python
wiki_data.head()
```

```python
context=[]
for idx, row in wiki_data.iterrows():
  filename=f'/content/wiki-documents-nsc/{row["article_id"]}.txt'
  f = open(filename, "r")
  context.append(f.read())
```

```python
wiki_data["context"]=context
```

```python
wiki_data.head()
```

```python
from lxml import etree
import xml.etree.ElementTree as ET
parser=etree.XMLParser(recover=True)
tree=ET.fromstring(wiki_data["context"][0],parser=parser)
print(tree.tag)
print(tree.attrib)
print(tree.text)
```

```python
def extract_text(body):
  tree=ET.fromstring(body,parser=parser)
  return tree.text
```

```python
wiki_data["context_text"]=wiki_data["context"].apply(extract_text)
```

```python
wiki_data.head()
```

```python
from pythainlp.tokenize import word_tokenize
wiki_data["context_text_wordsegged"]=wiki_data["context_text"].apply(lambda x: " ".join(word_tokenize(x,engine="newmm")))
```

```python
wiki_data["question_wordsegged"]=wiki_data["question"].apply(lambda x: " ".join(word_tokenize(x,engine="newmm")))
```

```python
wiki_data["answers_wordsegged"]=wiki_data["answer"].apply(lambda x: " ".join(word_tokenize(x,engine="newmm")))
```

```python
wiki_data.head()
```

```python
answer_starts=[]
for idx, row in wiki_data.iterrows():
  try:
    answer_start=row["context_text_wordsegged"].index(row["answers_wordsegged"])
  except:
    answer_start=-1
  answer_starts.append(answer_start)
```

```python
wiki_data["answer_starts"]=answer_starts
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
wiki_train_df, wiki_val_df = train_test_split(wiki_data, test_size=0.05, random_state=23)
```

```python
scg_train_df, scg_val_df = train_test_split(scg_data, test_size=0.05, random_state=23)
```

```python
train_data=[]
i=0
for idx, row in wiki_train_df.iterrows():
  doc={}
  doc["context"]=row["context_text_wordsegged"]
  doc["qas"]=[{"id":i, "is_impossible":row["answer_starts"]<0,"question":row["question_wordsegged"],"answers":[{"text":row["answers_wordsegged"],"answer_start":row["answer_starts"]}]}]
  train_data.append(doc)
  i=i+1
```

```python
val_data=[]
i=0
for idx, row in wiki_val_df.iterrows():
  doc={}
  doc["context"]=row["context_text_wordsegged"]
  doc["qas"]=[{"id":i, "is_impossible":row["answer_starts"]<0,"question":row["question_wordsegged"],"answers":[{"text":row["answers_wordsegged"],"answer_start":row["answer_starts"]}]}]
  val_data.append(doc)
  i=i+1
```

```python
pretrained="monsoon-nlp/bert-base-thai" #@param ["monsoon-nlp/bert-base-thai","bert-base"]
```

```python
name=pretrained
```

```python
import torch
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

model_args={
    "reprocess_input_data":True,
    "train_batch_size":32,
    'eval_batch_size' : 8,
    "num_train_epochs": 5,
    'evaluate_during_training' : True,
    'evaluate_during_training_verbose': True,
    "overwrite_output_dir": True}

model=QuestionAnsweringModel("bert",name, use_cuda=torch.cuda.is_available(), args=model_args)


model.train_model(train_data,eval_data=val_data, show_running_loss=True)

result,texts=model.eval_model(val_data)
```
