```python
import cv2
import json
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
for i in range(1,3001):
  path_gt ="/content/set1/"+str(i)+"/gt_text.json"
  path_img = "/content/set1/"+str(i)+"/image.png"
  f = open(path_gt)
  data = json.load(f)
  img = cv2.imread(path_img)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  for j in data:
    ymin = math.floor(j["rect"]["y"])
    ymax = math.ceil(j["rect"]["y"]+j["rect"]["height"])
    xmin = math.floor(j["rect"]["x"])
    xmax = math.ceil(j["rect"]["x"]+j["rect"]["width"])
    
    if ymax<img.shape[0] and ymax-ymin >0 and xmax-xmin>0:
      img_sub = img[ymin :ymax, xmin:xmax]
      '''
      plt.figure()
      plt.imshow(img_sub)
      plt.show()
      '''
      filename = j["parentId"]+"-"+j["text"]+"-"+j["style"]["fontFamily"]+"-"+j["style"]["fontSize"]+"-"+j["style"]["fontStyle"]+"-"+j["style"]["fontWeight"]
      print(filename)
      cv2.imwrite("/content/drive/MyDrive/MiniHackathon/DataCrop1/"+filename+".png", img_sub)
  f.close()
  print(i)
```
